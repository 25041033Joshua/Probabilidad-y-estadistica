## Precio Maximo

```python
# Mostrar el precio maximo de la base de datos de la columna price
autos2.price.max()
```

## Precio Minimo

```python
# Mostrar el precio minimo de la base de datos de la columna price
autos2.price.min()
```

## Media del precio de los autos


```python
#Utilizar la función mean que esta en la librería statics para sacar el promedio

from statistics import mean

media=autos2.price.mean()
print(media)
```

## Mediana del precio de los autos

```python
#Determinar el valor de la mediana con la funcion median
from statistics import median

mediana=autos2.price.median()
print(mediana)
```

## Moda de los precios de los autos

```python
#Sacar la moda de las calificaciones con la funcion mode
from statistics import mode
moda=autos2.price.mode()
print(moda)
```

# Desviación estándar


```python
desviación_estandar=autos2.price.std()
print(desviación_estandar)
```

# Estadisticos y Cuartiles

```python
estadisticos=autos2['price'].describe()
print(estadisticos)
```

# Regla de Sturges

```python
from pandas.core.api import DataFrame
import math
# Calcular el número de intervalos utilizando la regla de Sturges
n=len(autos2)
k=int(1+math.log2(n))
print("Valor de n=",n,"\nNúmero de intervalos",k)
```

# Histograma de precios de automoviles

```python
plt.hist(autos2['price'], bins=k, alpha=0.7,color='b', edgecolor = "black")
plt.axvline(media, color='blue', linestyle='dashed',linewidth=2, label='Media')
plt.axvline(mediana, color='green', linestyle='dashed',linewidth=2, label='Mediana')
plt.axvline(moda, color='purple', linestyle='dashed',linewidth=2, label='Moda')
plt.axvline(media-desviación_estandar, color='orange', linestyle='dashed',linewidth=2, label=f'desv. estandar={desviación_estandar:.2}')
plt.axvline(media+desviación_estandar, color='black', linestyle='dashed',linewidth=2)
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title("Histograma de los Precios de los Autos")
plt.legend()
plt.show()
```

### Gracias a este histograma tenemos una representación visual de como se distribuyen diferentes partes esenciales para el analisis de la información, como las area en donde hay mayor y menor concetración en los precios, pues facilita la identificación de las medidas de tendencia central asi como la desviación estandar ahorrando mucho tiempo para su analisis


# Diagramas de Caja y Bigote


```python

import matplotlib.pyplot as plt
from statistics import*
import pandas as pd
```

```python
sns.boxplot(x='body-style',y='price',data=autos2)
plt.title("Relación tipo de auto-precio")
plt.xlabel("Tipo de auto")
plt.ylabel("Precio")
```

### Este tipo de gráfica nos facilita comparar de forma directa los precios de los autos y sus tipos, pues gracias a esto no es necesario pasar tanto tiempo buscando y comparando uno por uno para determinar que opcion es mejor

```python
sns.boxplot(x='drive-wheels',y='price',data=autos2)
plt.title("Relación tipo de tracción de ruedas-precio")
plt.xlabel("Tracción de ruedas")
plt.ylabel("Precio")
```

### Al igual que con la grafica anterior podemos ver la comparación entre los tipos de ruedas de tracción y su respectivo precio, destacando que al seguir una representación visual permite una mejor toma de decisión

```python
sns.boxplot(x='engine-location',y='price',data=autos2)
plt.title("Relación de motores-precio")
plt.xlabel("Motor")
plt.ylabel("Precio")
```

### En esta grafica de caja y bigote podemos observar como la relación entre el tipo de motores traseros (rear) se ve muy reflejado en el precio de ser muy elevados, en cambio los motores delanteros (front) se encuentran precios más bajos a comparación de los traseros.
